```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/cars.csv')
df
```

```python
df.info()
```

```python
selected_columns = ['price_x', 'currency_x', 'city', 'production_year', 'engine_displacement_num',
                    'kilometrage_num', 'Marka', 'Model', 'Rəng',
                    'Sürətlər qutusu', 'Vəziyyəti', 'Yeni', 'Yerlərin sayı', 'Ötürücü']

df = df[selected_columns]
```

```python
def price_modify(df):
    if df['currency_x'] == '$':
        return df['price_x'] * 1.7
    elif df['currency_x'] == '€':
        return df['price_x'] * 2
    else:
        return df['price_x']

df.loc[:, 'price_x'] = df.apply(price_modify, axis=1)
df = df.drop('currency_x', axis = 1)
```

```python
def seat_modify(df):
    try:
        if len(df['Yerlərin sayı']) == 2:
            return int(df['Yerlərin sayı'][0])
        return int(df['Yerlərin sayı'])
    except:
        return df['Yerlərin sayı']

df['Yerlərin sayı'] = df.apply(seat_modify, axis=1)
```

```python
model_avg_prices = df.groupby('Model')['price_x'].mean()

df['Avg_model_price'] = df['Model'].map(model_avg_prices)
df = df.drop('Model', axis = 1)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.isna().sum()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price_x'])
plt.title('Box Plot of Price_x')
plt.xlabel('Price_x')

plt.show()
```

```python
df['price_x'].describe()
```

```python
df = df.loc[df['price_x'] <= 200000]
```

```python
X = df.drop('price_x', axis=1)
y = df['price_x'].copy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, random_state=42, test_size=0.30)
```

```python
X_num_train = X_train.select_dtypes(include=np.number)
X_cat_train = X_train.select_dtypes(include='object')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
```

```python
X_num_pipeline = Pipeline([
    ('Imputer', SimpleImputer(strategy='median')),
    ('Scaler', StandardScaler()),
    ('Poly', PolynomialFeatures(degree=2, include_bias=False)),
])

X_cat_pipeline = Pipeline([
    ('Imputer', SimpleImputer(strategy='most_frequent')),
    ('Encoder', OneHotEncoder(handle_unknown="ignore", sparse_output = False))
])

transformer = ColumnTransformer([
    ('X_num', X_num_pipeline, X_num_train.columns),
    ('X_cat', X_cat_pipeline, X_cat_train.columns)
], remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
full_pipeline_linear = Pipeline([
    ('preprocessing', transformer),
    ('linear', LinearRegression())
])

full_pipeline_linear.fit(X_train, y_train)
print(f"Train score: {full_pipeline_linear.score(X_train, y_train)}, Test score: {full_pipeline_linear.score(X_test, y_test)}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
full_pipeline_tree = Pipeline([
    ('preprocessing', transformer),
    ('tree', DecisionTreeRegressor())
])

full_pipeline_tree.fit(X_train, y_train)
print(f"Train score: {full_pipeline_tree.score(X_train, y_train)}, Test score: {full_pipeline_tree.score(X_test, y_test)}")
```

```python
print(f"Validation score: {full_pipeline_tree.score(X_val, y_val)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
full_pipeline_forest = Pipeline([
    ('preprocessing', transformer),
    ('forest', RandomForestRegressor(max_depth=14))
])

full_pipeline_forest.fit(X_train, y_train)
print(f"Train score: {full_pipeline_forest.score(X_train, y_train)}, Test score: {full_pipeline_forest.score(X_test, y_test)}")
```
